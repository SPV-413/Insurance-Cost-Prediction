```python
a.select_dtypes(include=['int','float'])
```

```python
a.rename({'Outcome':'Result'},axis=1,inplace=True)
a
```

```python
a['Glucose']=a['Glucose'].replace(0,a['Glucose'].mean())
a['Glucose']
```

```python
#replace corrupted datas that is zeroes in glucose,bmi,insulin,BP with mean value
s=['Glucose','BloodPressure','Insulin','BMI']
for i in s:
    a[i]=a[i].replace(0,a[i].mean())
    print(a[i])
```

```python
#find count of people between age 50 and 70 using condition and shape
a[(a['Age']>=50) & (a['Age']<=70)].shape[0] #1 way
```

```python
a[(a['Age']>=50) & (a['Age']<=70)].count() # another way
```

```python
#drop Insulin column without inplace
a=a.drop('Insulin',axis=1)
a
```

```python
#group the datas based on result and extract pregnancies and bmi column
list(a.groupby('Result')[['Pregnancies','BMI']])
```

```python
#find mean,max,count,min value for age column after groupby using result
a.groupby('Result').agg({'Age':['mean','max','count','min']})
```

```python
#group the data based on result and age ,result count values for each row
a.groupby(['Result','Age']).count()
```

```python
#create new column and categorize people based on bmi
#bmi<18 -underweight
#bmi between 18 and 25 -fit
#bmi>25-overweight
def new(BMI):
    if BMI<18:
        return 'Underweight'
    elif 18<=BMI<=25:
        return 'Fit'
    else:
        return 'Overweight'
a['Category']=a['BMI'].apply(new)
a
```

```python
#group the data based on result and age ,result count values for each columns
a.groupby(['Result','Age']).count()
```
